## 1. Import Libraries

```python
# Import Libraries
import pandas as pd
from datetime import datetime, timedelta
```

## 2. Data Loadiing

```python
df = pd.read_csv('data_raw.csv', encoding='iso-8859-1')
df
```

## 3. Data Cleaning

```python
df.info()
```

```python
# Drop Duplicate Row
df = df.drop_duplicates()
```

```python
# Fill Null Column
df['Postal_Code'] = df['Postal_Code'].fillna(5401.0)
```

```python
# Change Postal_Code Column Type from Float to String
df['Postal_Code'] = df['Postal_Code'].astype(int).astype(str).str.zfill(5)
```

```python
# Change data type from column Order_Date and Ship_Date to date time 
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True, errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], dayfirst=True, errors='coerce')
```

```python
# Change Column name to lower case
df.columns = df.columns.str.lower()
```

```python
df.info()
```

```python
# Save to CSV
df.to_csv('data_clean.csv')
```
